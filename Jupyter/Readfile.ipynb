```python
import pandas as pd 
file="file.json"
file2="file2.json"
import json
  
# Opening JSON file
f = open(file)
f2 = open(file2)
# returns JSON object as 
# a dictionary
data = json.load(f) 
data2 = json.load(f2) 
```

```python
df1=pd.json_normalize(data["hits"]) 
df2 = pd.json_normalize(data2["hits"]) 
```

```python
df = pd.concat([df1,df2])
```
